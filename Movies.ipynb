#### Collaborative Filtering

```python
ratings_data = ratings_data.drop('timestamp', axis=1)
```

```python
userInput = [
            {'title':'Toy Story ', 'rating':3.5},
            {'title':'Jumanji ', 'rating':2},
            {'title':"Pulp Fiction ", 'rating':5},
            {'title':'Akira ', 'rating':4.5}
         ] 
input_data = pd.DataFrame(userInput)
input_data
```

```python
Id = movies_data[movies_data['title'].isin(input_data['title'].tolist())]
input_data = pd.merge(Id, input_data)
input_data = input_data.drop('year', axis=1)
input_data
```

```python
userSubset = ratings_data[ratings_data['movieId'].isin(input_data['movieId'].tolist())]
userSubset
```

```python
userGroup = userSubset.groupby(['userId'])
```

```python
userGroup = sorted(userGroup,  key=lambda x: len(x[1]), reverse=True)
```

```python
userGroup
```

```python
userGroup = userGroup[0:150]
```

#### Using correlation analysis(Pearson) between user's data

```python
pearsonCorrelation = {}

for name, group in userGroup:
    group = group.sort_values(by='movieId')
    input_data = input_data.sort_values(by='movieId')
    nRatings = len(group)
    temp_df = input_data[input_data['movieId'].isin(group['movieId'].tolist())]
    tempRatingList = temp_df['rating'].tolist()
    tempGroupList = group['rating'].tolist()
    xx = sum([i**2 for i in tempRatingList]) - pow(sum(tempRatingList),2)/float(nRatings)
    yy = sum([i**2 for i in tempGroupList]) - pow(sum(tempGroupList),2)/float(nRatings)
    xy = sum( i*j for i, j in zip(tempRatingList, tempGroupList)) - sum(tempRatingList)*sum(tempGroupList)/float(nRatings)
    
    if xx != 0 and yy != 0:
        pearsonCorrelation[name] = xy/sqrt(xx*yy)
    else:
        pearsonCorrelation[name] = 0
```

```python
pearson = pd.DataFrame.from_dict(pearsonCorrelation, orient='index')
pearson.columns = ['similarity_Index']
pearson['userId'] = pearson.index
pearson.index = range(len(pearson))
pearson
```

```python
topUsers=pearson.sort_values(by='similarity_Index', ascending=False)[0:100]
topUsers
```

```python
topUsersRating=topUsers.merge(ratings_data, left_on='userId', right_on='userId', how='inner')
topUsersRating
```

```python
topUsersRating['weightedRating'] = topUsersRating['similarity_Index']*topUsersRating['rating']
topUsersRating
```

```python
changedTopUsersRating = topUsersRating.groupby('movieId').sum()[['similarity_Index','weightedRating']]
changedTopUsersRating.columns = ['sum_similarity_Index','sum_weighted_Rating']
changedTopUsersRating
```

```python
recommendation_data = pd.DataFrame()
recommendation_data['weighted average recommendation score'] = changedTopUsersRating['sum_weighted_Rating']/changedTopUsersRating['sum_similarity_Index']
recommendation_data['movieId'] = changedTopUsersRating.index
recommendation_data
```

```python
recommendation_data = recommendation_data.sort_values(by='weighted average recommendation score', ascending=False)
recommendation_data
```
